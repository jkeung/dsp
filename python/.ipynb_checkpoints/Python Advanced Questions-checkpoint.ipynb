```python
import pandas as pd
```

```python
df = pd.read_csv('faculty.csv', header = 0)
```

```python
df[' degree'] = df[' degree'].map(lambda x: x.replace('.','') )
```

```python
array = list(df[' degree'])
```

```python
collection = dict()
for items in array:
    for item in items.strip().split():
        if item in collection:
            collection[item] += 1
        else:
            collection[item] = 1
            
```

```python
from pprint import pprint
df = pd.read_csv('faculty.csv', header = 0)
df['name'] = df['name'].map(lambda x:x.split()[-1])
df.columns = ['name', 'degree', 'title', 'email']
df2 = df.set_index('name', append=True)
faculty_dict = df2.T.to_dict('defaultdict')
pprint(faculty_dict.items()[:])
```

```python
from pprint import pprint
df = pd.read_csv('faculty.csv', header = 0)
df['name'] = df['name'].map(lambda x:(x.split()[-1], x.split()[0]))
df.columns = ['name', 'degree', 'title', 'email']
df2 = df.set_index('name')
faculty_dict = df2.T.to_dict('list')
pprint([(key, faculty_dict[key]) for key in sorted(faculty_dict.keys())[:3]])
```
