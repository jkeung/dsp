```python
import pandas as pd
```

```python
df = pd.read_csv('faculty.csv', header = 0)
```

```python
df[' degree'] = df[' degree'].map(lambda x: x.replace('.','') )
```

```python
array = list(df[' degree'])
```

```python
collection = dict()
for items in array:
    for item in items.strip().split():
        if item in collection:
            collection[item] += 1
        else:
            collection[item] = 1
            
```
